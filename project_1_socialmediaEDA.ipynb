```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random as rnd
from datetime import datetime, timedelta
```

```python
# Function to generate random data dictionary
def generate_random_data(n):
    categories = [
        "Personal Updates",
        "News and Current Events",
        "Entertainment",
        "Humor and Memes",
        "Quotes and Inspirational Messages",
        "Technology",
        "Sports",
        "Food and Cooking",
        "Travel",
        "Health and Wellness",
        "Fashion and Style",
        "Gaming",
        "Social and Cultural Issues",
        "Product Reviews",
        "Educational Content",
        "Promotions and Marketing",
        "Behind the Scenes",
        "Polls and Surveys",
        "Ask Me Anything",
        "Weather Updates",
    ]

    data = {
        "Date": [(datetime.now() - timedelta(days=i)).strftime("%Y-%m-%d") for i in range(n)],
        "Category": [rnd.choice(categories) for _ in range(n)],
        "Likes": [rnd.randint(0, 10000) for _ in range(n)],
    }

    return data

# Generate random data with n = 1000
n = 1000
random_data = generate_random_data(n)
```

```python
# Reset display options to default (optional)
pd.reset_option("display.max_rows")
pd.reset_option("display.max_columns")

# Set display options to show all rows and columns
# pd.set_option("display.max_rows", None)
# pd.set_option("display.max_columns", None)

# Create a Pandas DataFrame from the random data
df = pd.DataFrame(random_data)

# Print the DataFrame head
print("DataFrame Head:")
print(df.head())

# Print the DataFrame information
print("\nDataFrame Information:")
print(df.info())

# Print the DataFrame description
print("\nDataFrame Description:")
print(df.describe())
```

```python
# Count unique categories from the "Category" column
unique_categories = df["Category"].value_counts()

# Print the count of unique categories
print(unique_categories)
```

```python
# Count likes for each category
likes_by_category = df.groupby('Category')['Likes'].sum()

# Print the result
print(likes_by_category)
```

```python
# Convert "Date" column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Convert "Likes" column to integers
df['Likes'] = df['Likes'].astype(int)
```

```python
# Print the DataFrame head
print("DataFrame Head:")
print(df.head())

# Print the DataFrame information
print("\nDataFrame Information:")
print(df.info())

# Print the DataFrame description
print("\nDataFrame Description:")
print(df.describe())
```
